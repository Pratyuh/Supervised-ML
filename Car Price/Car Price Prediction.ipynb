```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datasets = datasets.drop(['name'],axis=1)
```

```python
datasets = pd.get_dummies(datasets, columns =['fuel','seller_type','transmission','owner'],drop_first=True)
datasets.dtypes
```

```python
x = datasets.drop(['selling_price'],axis=1)
y = datasets['selling_price']
```

```python
x
```

```python
y
```

```python
print("Shape of the x :", x.shape)
print("Shape of the y :", y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
# lets print the shapes again 
print("Shape of the x Train :", x_train.shape)
print("Shape of the y Train :", y_train.shape)
print("Shape of the x Valid :", x_valid.shape)
print("Shape of the y Valid :", y_valid.shape)
```

```python
x
```

```python
from sklearn.metrics import r2_score,mean_squared_error
from math import sqrt
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(x_train,y_train)

#Predicting the Test set results
y_pred_rfr = rfr.predict(x_train)
score = r2_score(y_train,y_pred_rfr)
print("Score of Training:",100*score)
print("RMSE :" , np.sqrt(mean_squared_error(y_train,y_pred_rfr)))

y_test_pred_rfr = rfr.predict(x_valid)
score = r2_score(y_valid,y_test_pred_rfr)
print("Score of Testing:",100*score)
print("RMSE : " , np.sqrt(mean_squared_error(y_valid,y_test_pred_rfr)))
```

```python
accur = rfr.score(x_valid,y_valid)
accur
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=1000, learning_rate=0.05)
xgb.fit(x_train,y_train)

#Predicting the Test set results
y_pred_rfr = xgb.predict(x_train)
score = r2_score(y_train,y_pred_rfr)
print("Score of Training:",100*score)
print("RMSE :" , np.sqrt(mean_squared_error(y_train,y_pred_rfr)))

y_test_pred_rfr = xgb.predict(x_valid)
score = r2_score(y_valid,y_test_pred_rfr)
print("Score of Testing:",100*score)
print("RMSE : " , np.sqrt(mean_squared_error(y_valid,y_test_pred_rfr)))
```
